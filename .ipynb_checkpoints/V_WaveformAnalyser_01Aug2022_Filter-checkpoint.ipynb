```python
import lecroyparser
import csv
import matplotlib.pyplot as plt
import numpy as np
from scipy import optimize
from scipy import signal
Polynomial = np.polynomial.Polynomial
%matplotlib inline
plt.rcParams['figure.dpi'] = 250
plt.rcParams['savefig.dpi'] = 250
```

### Gaussian Function: gaus()

```python
#Used for fitting results
def gaus(x, A, mean, sigma):
    """
    Gaussian function with A as the normalisation/amplitude, mean and sigma. 
    """
    return A * np.exp(-(x - mean)**2.0 / (2 * sigma**2))
```

### Threshold Crossing Detection: thresh_interpolated()

```python
def thresh_interpolated(X, taxis, threshold, no_points=4, print_=False):
    """
    Returns the interpolated threshold crossing (negative going) time/index.
    no_points is the number of points by two to use in the interpolation
    """
    thresh_ind=0
    for a in X:
        #this loop ends when a[thresh_ind] crosses the threshold
        if a<threshold: break
        thresh_ind+=1

    if(thresh_ind!=len(X) and thresh_ind>no_points):
        xtofit=X[thresh_ind-no_points:thresh_ind+no_points]
        taxisfit=taxis[thresh_ind-no_points:thresh_ind+no_points]
        cmin, cmax = min(xtofit), max(xtofit)
        pfit, stats = Polynomial.fit(taxisfit, xtofit, 1, full=True, window=(cmin, cmax), domain=(cmin, cmax))
        A0, m = pfit
        thresh_time=(threshold-A0)/m
        if(print_):
            print(xtofit)
            print(taxisfit)
            print('\n A0: '+str(A0)+'   b: '+str(m))
            print('\n thresh_time: ' + str(thresh_time))
            print('\n stats:')
            print(stats)
    else: thresh_time=-999999999999
    if(print_):
        if(thresh_ind==len(X) or thresh_ind<=no_points): print ('didn\'t cross threshold')
        else:
            print('X[t]  X[t-1]  X[t-2]: '+ str(X[thresh_ind])+'  '+str(X[thresh_ind-1])+'  '+str(X[thresh_ind-2]))
            print('thresh_ind: ' + str(thresh_ind))
            print('thresh_time: ' + str(thresh_time))

    return thresh_time
```

### Baseline Maximum CFD: baseline_max_cfd()

```python
def baseline_max_cfd(baseline_pts=100, print_=False):
    """returns tuple (Xbaseline, Xmax, threshXcfd)"""
    
    #############################################
    #Calculating Baseline
    #############################################
    if(print_): print('\n*********Calculating Baseline*********')
    XbaseLine = sum(X[:baseline_pts])/baseline_pts
    if(print_): print('baseline = ', baseLine)   
    
    #############################################
    #Calculating Maximum
    #############################################
    ##### Actually minimum, this code is for negative going signals
    if(print_): print('\n*********Calculating Maximum*********')
    Xmax=min(X)
    if(print_): print('Xmax = ', Xmax)

    #############################################
    #Constant Fraction Discrimination Thresholds
    #############################################
    if(print_): print('\n*********Calculating CFD Thresholds*********')
    threshXcfd = XbaseLine-cfd*(XbaseLine-Xmax)
    if(print_): print('threshXcfd: ', threshXcfd)
    
```

### Run Settings

```python
#Settings
no_samples=1000
folder1chX='../generated_files/genX'#folder and file name (prefix) for channel X
folder2chY='../generated_files/genY'#folder and file name (prefix) for channel Y
binary_=False #Keep true if the input files are in LeCroy binary format, False if Lecroy CSV format
plot_=False
print_=False


#########################Filter Settings
filter_=False #set true to enable filter
fc=0.025


############################# simple threshold settings
##################### threshX, threshY are for thesholds for channels X and Y respectively 
#############################
threshX=-0.2
threshY=-0.2

############################# Constant Fraction Descriminator Settings
##################### threshX_min, threshY_min are the minimum allowed thresholds
##################### if threshXcfd-baseline (or threshYcfd-baseline) is </>(polarity) threshX_min (or threshY_min)
#####################  ...the event will be ignored
#############################
cfd=0.5
baseline_pts=200 #number of points to bee used to calculate the baseline
threshX_min=0.005 #minimum threshold above or below (corresponding to +ve or -ve polarity) the baseline for channel X
threshY_min=0.005 #minimum threshold above or below (corresponding to +ve or -ve polarity) the baseline for channel Y

############################# plot settings
#############################
xlim_low=-0.1
xlim_high=+0.1
ylim_low=-0.1
ylim_high=+0.1

#################
#######
#######
#######TBD:option for CFD vs normal, option for inverted pulses
#######Automate the xlim and ylim for plotting
#######
#######
#################


```

### Main Analysis Loop

```python
#this is the edited code......... 28 september 2022

############################# DATA HOLDERS
#############################
time_diff=[] #time differences with simple threshold crossing
time_diff_int=[] #time differences with interpolated threshold crossing
time_diff_int_cfd=[] #time differences with interpolated threshold crossing with cfd
baselineX=[]
baselineY=[]
threshXcfd=[]
threshYcfd=[]
X_Max=[]
Y_Max=[]
#to add: maximum-baseline

for sample_no in range(0, no_samples): 
    if(sample_no%100==0): print(sample_no)
    if(print_): print('\n\n\n SAMPLE_NO: '+str(sample_no))
    timestampX=[]
    X=[]
    timestampY=[]
    Y=[]
    #Reading folder1chX 
    if(binary_): filename=folder1chX+str(sample_no).zfill(5)+'.trc'
    else: filename=folder1chX+str(sample_no).zfill(5)+'.csv'
    if(print_): print(filename)
        
    if(binary_):
        data = lecroyparser.ScopeData(filename)
        timestampX=data.x
        X=data.y
    else:
        with open(filename, "rt") as infile:
            file_read = csv.reader(infile)
            for _ in range(5): next(file_read, None) #ignore the five lines of header
            for row in file_read:
                timestampX.append(float(row[0]))
                X.append(float(row[1]))
    #folder2chY
    if(binary_): filename=folder2chY+str(sample_no).zfill(5)+'.trc'
    else: filename=folder2chY+str(sample_no).zfill(5)+'.csv'
    if(print_): print(filename)
    if(binary_):
        data = lecroyparser.ScopeData(filename)
        timestampY=data.x
        Y=data.y
    else:
        with open(filename, "rt") as infile:
            file_read = csv.reader(infile)
            for _ in range(5): next(file_read, None) #ignore the five lines of header
            for row in file_read:
                timestampY.append(float(row[0]))
                Y.append(float(row[1]))
                
    #############################################
    #Lowpass Filtering
    #############################################
    if(filter_==True):
        b, a = signal.butter(10, fc, 'low')
        X = signal.filtfilt(b, a, X)
        Y = signal.filtfilt(b, a, Y)
    
###########here was baseline, max, cfd code. use function from now on.
    #############################################
    #Plotting
    #############################################
    if(plot_):
        if(print_): print('\n*********Plotting*********')
        fig, ax1 = plt.subplots()
        #plotting channelX
        color = 'tab:red'
        ax1.set_xlabel('time (s)')
        ax1.set_ylabel('Channel X', color=color)       
        ax1.tick_params(axis='y', labelcolor=color)
        ax1.plot(timestampX, X, color=color)
        ax2 = ax1.twinx()
        #plotting channelY
        color = 'tab:blue'
        ax2.set_ylabel('Channel Y', color=color)  # we already handled the x-label with ax1
        ax2.plot(timestampY, Y, color=color)
        ax2.tick_params(axis='y', labelcolor=color)
        plt.title('sample_no'+str(sample_no))
        fig.tight_layout()
        ax1.grid(True)
        ax2.grid(True)
        ax1.set_ylim([xlim_low, xlim_high])
        ax2.set_ylim([ylim_low, ylim_high])
        
        #plotting a line to mark the threshold
        ax1.plot(timestampX, [threshXcfd[len(threshXcfd)-1]] * len(X), color='r', linewidth=0.7, linestyle='dashed', label=' ')
        ax2.plot(timestampY, [threshYcfd[len(threshYcfd)-1]] * len(Y), color='b', linewidth=0.7, linestyle='dashed', label=' ')
        ax1.text(timestampX[100], -0.04, 'threshXcfd:'+"{:.5f}".format(threshXcfd[len(threshXcfd)-1])+ '\nthreshYcfd:'+"{:.5f}".format(threshYcfd[len(threshYcfd)-1]), fontsize = 12)

        #saving the plots
        plt.savefig('./plots/sample_no'+str(sample_no)+'.jpeg')
        if(print_): print('Plot Saved')
        plt.close()

    #############################################
    ##Simple threshold crossing
    #############################################
    if(print_): print('\n*********Simple Threshold Crossing*********')
        
    thresh_time_index_X, thresh_time_index_Y = 0, 0
    slopeX, slopeY, timing_diff_in = 0.0, 0.0, 0.0
    #channel X
    thresh_time_index_X=0;
    for a in X:
        if a<threshX: break
        thresh_time_index_X+=1
    
    #channel Y
    thresh_time_index_Y=0;
    for a in Y:
        if a<threshY: break
        thresh_time_index_Y+=1    
    if(thresh_time_index_Y!=len(Y) and thresh_time_index_X!=len(X)):
        timing_diff_in = thresh_time_index_Y-thresh_time_index_X
        time_diff.append(timing_diff_in)
    
    #############################################
    ##Interpolating Threshold Crossing
    #############################################
    if(print_): print('\n*********Interpolating Threshold Crossing*********')
        
    #channel X
    thresh_time_X=1e12*thresh_interpolated(X, timestampX, threshX, 4, print_)
    #channel Y
    thresh_time_Y=1e12*thresh_interpolated(Y, timestampY, threshY, 4, print_)

    if(thresh_time_Y!=-999999999999 and thresh_time_X!=-999999999999):
        timing_diff_in = thresh_time_Y-thresh_time_X
        time_diff_int.append(timing_diff_in)
        if(print_):
            print('timing difference: '+str(timing_diff_in))

    #############################################
    ##Interpolating Threshold Crossing with CFD
    #############################################
    if(print_):
        print('\n*********Interpolating Threshold Crossing with CFD*********')
        print('baselineX-threshXcfd: '+str(baselineX[len(baselineX)-1]-threshXcfd[len(threshXcfd)-1]))
        print('baselineY-threshYcfd: '+str(baselineY[len(baselineY)-1]-threshYcfd[len(threshYcfd)-1]))
    
    if(baselineX[len(baselineX)-1]-threshXcfd[len(threshXcfd)-1]>threshX_min and baselineY[len(baselineY)-1]-threshYcfd[len(threshYcfd)-1]>threshY_min):
        #channel X
        thresh_time_X=1e12*thresh_interpolated(X, timestampX, threshXcfd[len(threshXcfd)-1], 4, print_)
        #channel Y
        thresh_time_Y=1e12*thresh_interpolated(Y, timestampY, threshYcfd[len(threshYcfd)-1], 4, print_)

        timing_diff_in = thresh_time_Y-thresh_time_X
        time_diff_int_cfd.append(timing_diff_in)
        if(print_):
            print('timing difference: '+str(timing_diff_in)+'\n\n')
    ########NOTE THAT SOME VARIABLES REUSED: thresh_time_X, thresh_time_Y, timing_diff_in
```
