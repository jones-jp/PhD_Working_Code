```python

binary_=True
if(binary_==True):
    folder_name='../Data_RPC_PositionFromTiming/Noise_Measurements'
    filename=folder_name+'/FEndOn_002us_10mV_1M_40GSps_C2--ble--00009.trc'
else:
    filename='../generated_files/genX'+str(10).zfill(5)+'.csv'
#FEndOff_002us_05mV_50ohm_40GSps_C2--ble--00006.trc
#FEndOff_002us_05mV_50ohm_40GSps_C3--ble--00006.trc
#FEndOn_002us_05mV_50ohm_40GSps_C3--ble--00008.trc
#FEndOn_002us_05mV_50ohm_40GSps_C2--ble--00007.trc
#FEndOn_002us_10mV_1M_40GSps_C2--ble--00009.trc

if(binary_):
    data = lecroyparser.ScopeData(filename)
    timestampX=data.x
    X=data.y
else:
    timestampXl, Xl = [],[] #l for python list
    with open(filename, "rt") as infile:
        file_read = csv.reader(infile)
        for _ in range(5): next(file_read, None) #ignore the five lines of header
        for row in file_read:
            timestampXl.append(float(row[0]))
            Xl.append(float(row[1]))
    X=np.array(Xl)
    timestampX=np.array(timestampXl)

print(len(X))
print('rms = ' + str(np.mean(X*X)))

fs=40e9
f, Pxx_den = signal.welch(X, fs, nperseg=1024)

plt.semilogy(f, Pxx_den)

#plt.ylim([0.5e-3, 1])

plt.xlabel('frequency [Hz]')

plt.ylabel('PSD [V**2/Hz]')

plt.show()
```
